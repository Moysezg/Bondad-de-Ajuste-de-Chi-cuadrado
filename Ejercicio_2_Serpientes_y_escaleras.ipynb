```python
lista=[]
lanzamientos=0
suma=0
promedio=0
import random
for i in range(1,10000):
  while suma<20:
    n = (int(random.uniform(1, 7)))
    suma=suma+n
    #print("dado:",n)
    if suma==3:
        suma=suma+8
        #print("sube a 11")
    if suma==15:
        suma=suma+4
        #print("sube a 19")
    if suma==13:
        suma=suma-9
        #print("baja a 4")
    if suma==17:
        suma==suma-7 
        #print("baja a 10")
    #print("estas en la casilla:",suma)
    lanzamientos=lanzamientos+1
  #print("Juego",i,"Numero de lanzamientos:",lanzamientos)
  promedio=lanzamientos+promedio
  suma=0
  lanzamientos=0
  promedioa=round(promedio/i,5)
  lista.append(promedioa)
#print(lista)

import matplotlib.pyplot as plt

plt.plot(lista)   # Dibuja el gráfico
plt.title("Serpientes y escaleras")   # Establece el título del gráfico
plt.xlabel("juegos")   # Establece el título del eje x
plt.ylabel("lanzamientos")   # Establece el título del eje y
#print(lista_bp)
```
